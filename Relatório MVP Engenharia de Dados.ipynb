# 1. Contexto

Em 2015, líderes de 193 países adotaram um plano global para combater a fome e a pobreza, bem como para promover o desenvolvimento sustentável para as gerações atuais e futuras, sem deixar ninguém para trás. A partir das lições aprendidas através da implementação dos Objetivos do Milênio (2000-2015), a Agenda das Nações Unidas para o Desenvolvimento Sustentável, aprovada durante a conferência Rio+20 e adotada através da Resolução A/RES/70/1 da Assembleia Geral das Nações Unidas, é um plano de ação que visa a fortalecer a paz universal em maior liberdade, erradicar a pobreza em todas as suas formas e dimensões, incluindo a pobreza extrema, e a tomar as medidas ousadas e transformadoras que são urgentemente necessárias para colocar o mundo em um caminho sustentável e resiliente. 

A Agenda apresenta 17 Objetivos de Desenvolvimento Sustentável (ODS) e 169 metas integrais e individisíveis:

O processo de implementação dos Objetivos, as responsabilidades da Agenda 2030 são globais e compartilhadas por governos, empresas, sociedade civil e indivíduos. A nível programático, a resolução 72/279 da Assembleia Geral eleva o Marco de Cooperação das Nações Unidas para o Desenvolvimento Sustentável (UNSDCF) como “o instrumento mais importante para o planejamento e a implementação das atividades de desenvolvimento da ONU em nível nacional, em apoio à implementação da Agenda 2030 para o Desenvolvimento Sustentável (Agenda 2030)”. Adotado entre o Sistema das Nações Unidas em um determinado país e seu Estado-membro, o Marco de Cooperação orienta todo o ciclo programático de atuação das agências, fundos e programas da ONU em um determinado país para impulsionar o planejamento, a implementação, o monitoramento, a prestação de contas e a avaliação do apoio coletivo da ONU para alcançar a Agenda 2030. O UNSDCF também apoia a configuração do fluxo de recursos e financiamento direcionados ao desenvolvimento sustentável, bem como o estabelecimento de parcerias estratégicas e efetivas.

# 4. Extração dos dados
Como mencionado anteriormente, a base de dados gerada a partir do UN.INFO é pública e está disponibilizada no [Portal de Dados do Grupo de Desenvolvimento Sustentável da ONU (UNSDG)](https://uninfo.org/data-explorer/cooperation-framework), na aba "Data Explorer".  A base de dados é disponibilizada em formato **.csv**. Os dados, em geral, são atualizados na plataforma trimestralmente, ou a depender da dinâmica de governança de dados do Grupo de Monitoramento, Avaliação e Aprendizagem (MEL) da ONU em cada país.

Na linguagem de MEL das Nações Unidas e de acordo com a teoria de mudança do Marco de Cooperação, _outcomes_ correspondem aos resultados esperados no âmbito do Marco de Cooperação, conforme apresentado na seção 2. Por sua vez, _outputs_ correspondem aos diferentes conjuntos de intervenções (ou produtos) através das quais o Sistema ONU agrega valor e que poderiam ajudar o país a superar os principais desafios de desenvolvimento identificados, acelerando o seu progresso na realização dos ODS. Por fim, um _sub-output_ ao nível mais operacional da teoria da mudança, representando ações individuais ou coletivas das agências, fundos e programas que, agregadas, compõem o conjunto de outputs.

O Portal de Dados permite a extração de dados em duas camadas de análise: a nível de relatórios de sub-outputs e a nível de indicadores. Para fins deste exercício, será analisada a **base de dados de sub-outputs** por ser a camada mais granular possível dos dados, trazendo informações financeiras e orçamentárias das ações de cada AFP, bem como informações desagregadas sobre beneficiários, natureza das ações e, mais importante para este exercício, os ODS relacionados.

O processo de exportação é simples, bastando apenas selecionar, na aba "Data Explorer", o país e o plano de desenvolvimento de análise: neste caso, o Marco de Cooperação das Nações Unidas para o Desenvolvimento Sustentável.

## Refinamento PRATA

```python
#Definir o uso padrão do ESQUEMA PRATA
spark.sql("USE CATALOG mvp")
spark.sql("USE SCHEMA silver")
```

A seguir, irei formatar alguns tipos de dados (Strategic Priority Code, datas e valores)

```python
from pyspark.sql.functions import when, expr, col, try_to_date

df_bronze = spark.table("mvp.bronze.br_jwp")

columns_int = ["Strategic_priority_code"]
columns_date = ["Start_date", "End_date"]
columns_float = [
    "Total_required_resources", "Total_available_resources",
    "2023_Required", "2023_Available", "2023_Expenditure",
    "2024_Required", "2024_Available", "2024_Expenditure",
    "2025_Required", "2025_Available", "2025_Expenditure",
    "2026_Required", "2026_Available",
    "2027_Required", "2027_Available"
]

df_casted = df_bronze
for col_name in columns_int:
    df_casted = df_casted.withColumn(
        col_name,
        when(
            expr(f"try_cast({col_name} as int)").isNotNull(),
            expr(f"try_cast({col_name} as int)")
        ).otherwise(0)
    )

for col_name in columns_date:
    df_casted = df_casted.withColumn(
        col_name,
        try_to_date(col(col_name), "yyyy-MM-dd")
    )

for col_name in columns_float:
    df_casted = df_casted.withColumn(
        col_name,
        when(
            expr(f"try_cast({col_name} as float)").isNotNull(),
            expr(f"try_cast({col_name} as float)")
        ).otherwise(0.00)
    )

df_casted.write.format("delta") \
    .option("mergeSchema", "true") \
    .mode("overwrite") \
    .saveAsTable("mvp.silver.br_jwp_silver")

display(df_casted.limit(5))
```

## Refinamento OURO
Nesta última etapa de transformação dos dados, criarei diferentes tabelas FATO e DIMENSÃO.

Como apresentado na seção 2, esta base de dados apresenta algumas colunas com estrutura multi-valorada, dificultando analises desagregadas por ODS, beneficiarios ou parceiros (doadores). Ao fazer uma busca simples, como exemplificarei abaixo, a base dados apenas consegue mostrar de forma agregada quais ODS são contemplados por um projeto, como se fossem uma unidade. Este problema se repete para beneficiarios e parceiros contribuintes.

Para superar este problema, optou-se por modelar os dados dentro de um **esquema de constelacão de fatos**. Esta modelagem foi especialmente selecionada pelo fato de o processo analítico exigir múltiplos processos de negócios relacionaods, mas que não podem ser compartilhadús por um único fato. Em outras palavras, faremos um split e explode das informaçōes multivaloradas, além de proporcionar informaçōes financeiras entre cada nova linha. Com isso, transformaremos cada objeto multivalorado em uma tabela fato distinta, e tais tabelas fato compartilharão atributos em comum. Em razão do meu conhecimnento (ainda) limitado sobre as infinitas possibilidades dá modelagem e analise de dados, acredito que esta modelagem me ajudará a resolver o problema.

O desafio está em conseguir fracionar e distribuir os recursos orçamentários de acordo com os ODS, parceiro e beneficiários. Para tanto, serão criadas as seguintes tabelas:

**Tabela fato (1):** expenditure x **SDG** x year

**Tabela fato (2):** expenditure x **partner** x year

**Tabela fato (3):** expenditure x **beneficiaries** x year

**Dimensōes compartilhadas:** 
- dim_agency;
- dim_beneficiaries;
- dim_contributing_partners;

```python
from pyspark.sql.functions import expr, col, ltrim, split, explode, size, to_timestamp

df_silver = spark.table("mvp.silver.br_jwp_silver").select("ID", "SDG_Goals", "2023_Expenditure", "2024_Expenditure", "2025_Expenditure")
df_sdg = spark.table("mvp.gold.dim_sdg").select("sdg_sk", "ID", "SDG")
df_tempo_ts = spark.table("mvp.gold.dim_tempo_ts").select("SK_tempo", "Year")

# Explode SDGs and count how many per row
df_exploded = (
    df_silver
    .withColumn("SDG_list", split(col("SDG_Goals"), ";"))
    .withColumn("num_sdgs", size(col("SDG_list")))
    .withColumn("SDG", explode(col("SDG_list")))
    .withColumn("SDG", ltrim(col("SDG")))
)

# Distribute expenditures equally across SDGs
df_distributed = (
    df_exploded
    .withColumn("2023_Expenditure", (col("2023_Expenditure") / col("num_sdgs")).cast("decimal(18,2)"))
    .withColumn("2024_Expenditure", (col("2024_Expenditure") / col("num_sdgs")).cast("decimal(18,2)"))
    .withColumn("2025_Expenditure", (col("2025_Expenditure") / col("num_sdgs")).cast("decimal(18,2)"))
)

df_joined = (
    df_distributed
    .join(df_sdg, on=["ID", "SDG"], how="left")
)

# Unpivot expenditures by year, ensuring Year is timestamp
df_long = (
    df_joined
    .select(
        "ID", "sdg_sk",
        expr("2023_Expenditure as Expenditure"),
        to_timestamp(expr("'2023-01-01'"), "yyyy-MM-dd").alias("Year")
    )
    .unionByName(
        df_joined.select(
            "ID", "sdg_sk",
            expr("2024_Expenditure as Expenditure"),
            to_timestamp(expr("'2024-01-01'"), "yyyy-MM-dd").alias("Year")
        )
    )
    .unionByName(
        df_joined.select(
            "ID", "sdg_sk",
            expr("2025_Expenditure as Expenditure"),
            to_timestamp(expr("'2025-01-01'"), "yyyy-MM-dd").alias("Year")
        )
    )
)

df_final = (
    df_long
    .join(df_tempo_ts, df_long.Year == df_tempo_ts.Year, how="left")
    .select(
        "ID", "sdg_sk", "SK_tempo", "Expenditure"
    )
)

df_final.write.mode("overwrite").saveAsTable("mvp.gold.fact_expenditure_sdg_year_normalized")
```

## Primeira pergunta de análise de negócio

**_Entre 2023 e 2025, para quais Objetivos de Desenvolvimento Sustentável tem sido direcionado o financiamento recebido pelas agências, fundos e programas da ONU no Brasil?_**

```python
%sql
SELECT
  d.SDG,
  SUM(CASE WHEN t.Year = 2023 THEN f.Expenditure ELSE 0 END) AS Expenditure_2023,
  SUM(CASE WHEN t.Year = 2024 THEN f.Expenditure ELSE 0 END) AS Expenditure_2024,
  SUM(CASE WHEN t.Year = 2025 THEN f.Expenditure ELSE 0 END) AS Expenditure_2025,
  SUM(f.Expenditure) AS Total_Expenditure
FROM mvp.gold.fact_expenditure_sdg_year_normalized f
JOIN mvp.gold.dim_sdg d
  ON f.sdg_sk = d.sdg_sk
JOIN mvp.gold.dim_tempo t
  ON f.sk_tempo = t.sk_tempo
GROUP BY d.SDG
ORDER BY Total_Expenditure DESC;
```

A liderança foi seguida pelos ODSx 17 " Partnership for the Goals" (USD 30.348.492,82) e ODS 16 "Peace and Justice - Strong Institutions" ( (USD) 25.407.434,81).

Já o ODS 7, relacionado à disponibilização e garantia equitativa de energia limpa e acessível teve o menor volume de investimentos executados, somando USD 583.096,25.

O query abaixo, ilustrado pela Figura 7, mostra a constância da predominância do ODS 4 frente aos demais Objetivos ao longo de cada ano. Ainda que haja oscilaçōes entre os demais 16 Objetivos ao longo dos anos, o Objetivo de Educação de Qualidade permanece na liderança.

```python
%sql
WITH ranked_expenditure AS (
  SELECT
    t.Year,
    d.SDG,
    SUM(f.Expenditure) AS Total_Expenditure,
    ROW_NUMBER() OVER (
      PARTITION BY t.Year
      ORDER BY SUM(f.Expenditure) DESC
    ) AS rn
  FROM mvp.gold.fact_expenditure_sdg_year_normalized f
  JOIN mvp.gold.dim_sdg d
    ON f.sdg_sk = d.sdg_sk
  JOIN mvp.gold.dim_tempo t
    ON f.sk_tempo = t.sk_tempo
  WHERE t.Year IN (2023, 2024, 2025)
  GROUP BY t.Year, d.SDG
)
SELECT
  Year,
  SDG,
  Total_Expenditure
FROM ranked_expenditure
WHERE rn <= 3
ORDER BY Year, Total_Expenditure DESC;
```

## Segunda pergunta de análise de negócio

**_O quanto cada parceiro contribuinte (doador) ajudou a promover a Agenda 2030 no período reportado? Quem são os principais contribuintes no Brasil?_**

A primeira análise abaixo nos mostra que, entre 2023 e 2025, os recursos implementados pela ONU no Brasil vieram dos TOP 5 parceiros contribuintes:
1) **UNICEF Other Resources**, no total de USD 49.721.292 executados, ou 20,67% de todos os recursos implementados.
2) **The US Government Department of State's Bureu of Population, Refugees and Migration**, no total de USD 18.645.781 executados, ou 7,76% de todos os recursos implementados.
3) **Agência Brasileira de Cooperação (ABC)**, no total de USD 18.292.944 executados, ou 7,61% de todos os recursos implementados.
4) **Rede Globo**, no total de USD 10.664.045 executados, ou 4,43% de todos os recursos implementados.
5) **Governo do Estado de Rondônia**, no total de USD 10.190.534 executados, ou 4,43% de todos os recursos implementados.

O UNICEF, por se tratar de um Fundo da ONU voltado à promoção dos direitos das crianças e adolescentes, possui um mandato único de captação de recursos, permitindo receber doaçōes de grandess parceiros do setor privado (_major donors_), sociedades filantrópicas, fundaçōes e pessoas físicas, do Brasil e de outros paises. A nomeclatura " UNICEF Other Resources"  se refere, portanto, a este conglomerado de doaçōes, sendo impossivel, a partir desta base de dados, especificar a origem especifica de cada recurso.


```python
%sql
SELECT
  Partner,
  SUM(Expenditure) AS Total_Expenditure,
  ROUND(SUM(Expenditure) / SUM(SUM(Expenditure)) OVER () * 100, 2) AS Percentage_of_Total
FROM mvp.gold.vw_expenditure_partner_year
WHERE Year IN (2023, 2024, 2025)
GROUP BY Partner
ORDER BY Total_Expenditure DESC
LIMIT 5;
```

O próximo query abaixo ilustra os TOP 3 parceiros contribuintes por ano, salvaguardadas as observaçōes feitas acima. Observa-se uma predominância dos recursos do UNICEF nos três anos e o surgimento da Itaipu Binacional em 2025.

```python
%sql
WITH ranked_partners AS (
  SELECT
    Year,
    Partner,
    SUM(Expenditure) AS Total_Expenditure,
    ROW_NUMBER() OVER (
      PARTITION BY Year
      ORDER BY SUM(Expenditure) DESC
    ) AS rn
  FROM mvp.gold.vw_expenditure_partner_year
  GROUP BY Year, Partner
)
SELECT
  Year,
  Partner,
  Total_Expenditure
FROM ranked_partners
WHERE rn <= 3
ORDER BY Year, Total_Expenditure DESC;
```

Embora os dados mostrem o "UNICEF Other Resources" como TOP 1 parceiro, **esta informação está equivocada**, em razão dos desafios ainda encontrado na versão beta do portal de dados e metodologias de coleta de dados.

O Governo brasileiro possui um modelo particular de cooperação técnica internacional com a ONU, regido pelo Decreto 5.151/2004. Por este Decreto,  acordos de cooperacao entre organismos internacionais e a Administração Publica e cujos recursos advêm da própria Administração (self-funded projects) possuem governança tripartite entre a AFP, a Agência Brasileira de Cooperação (ABC) e a entidade do governo que receberá a cooperação técnica. Em razão de a base de dados estar em versão beta, não há uma padronização de reporting, de forma que algumas AFPs apontam a ABC como parceira contribuinte e outras, a terceira parte. Portanto, os recursos apontados como advindos da ABC estão subestimados e não retratam a realidade. 

Para resolver esta questão, proponho um agregação de todos os recursos oriundos da ABC ou cujo parceiro registrado no Sistema comece com “#BR” ou "Brazilian", maneiras pelas quais a plataforma identifica parceiros governamentais brasileiros. Também fiz uma retirada manual de alguns dados marcados no sistema com #BR, mas que nao se encaixam neste quesito (recursos são oriundos de outras fontes, como parcerias público-privadas, emendas parlamentares ou termos de ajuste de conduta - TAC). Esta solução ainda não representa a realidade total, mas é uma excelente aproximação.

```python
%sql
SELECT
  cp.Partner,
  SUM(f.Expenditure) AS Total_Expenditure
FROM mvp.gold.fact_expenditure_partner_year f
JOIN mvp.gold.dim_contributing_partners cp
  ON f.sk_partner = cp.sk_partner
WHERE (
    cp.Partner = 'ABC - Brazilian Cooperation Agency'
    OR cp.Partner LIKE '#BR%'
    OR cp.Partner LIKE 'Brazilian%'
  )
  AND cp.Partner NOT IN (
    '#BR Rede Globo',
    '#BR Itaipu Binacional',
    '#BR Serviço Social da Indústria',
    'Fundação #BR Renova',
    '#BR UNHCR PSP Brazil',
    '#BR Gerdau',
    '#BR Petrobrás'
  )
GROUP BY cp.Partner

UNION ALL

SELECT
  'Total' AS Partner,
  SUM(f.Expenditure) AS Total_Expenditure
FROM mvp.gold.fact_expenditure_partner_year f
JOIN mvp.gold.dim_contributing_partners cp
  ON f.sk_partner = cp.sk_partner
WHERE (
    cp.Partner = 'ABC - Brazilian Cooperation Agency'
    OR cp.Partner LIKE '#BR%'
    OR cp.Partner LIKE 'Brazilian%'
  )
  AND cp.Partner NOT IN (
    '#BR Rede Globo',
    '#BR Itaipu Binacional',
    '#BR Serviço Social da Indústria',
    'Fundação #BR Renova',
    '#BR UNHCR PSP Brazil',
    '#BR Gerdau',
    '#BR Petrobrás'
  )

ORDER BY 
  CASE WHEN Partner = 'Total' THEN 1 ELSE 0 END,
  Total_Expenditure DESC;
```

Por esse query, podemos inferir que a ONU e a ABC possuem acordos de cooperação técnica tripartite com PELO MENOS 54 órgãos governamentais. No total, **as iniciativas implementadas com recursos públicos através de cooperação internacional somam, pelo menos, USD 104.519.607, claramente se configurando como a principal origem de recursos implementados pela ONU no Brasil - 47,43%**.

```python
%sql
SELECT
  CASE 
    WHEN cp.Partner = 'ABC - Brazilian Cooperation Agency' 
      OR cp.Partner LIKE '#BR%' 
      OR cp.Partner LIKE 'Brazilian%' 
      THEN 'Self-funded partners'
    ELSE cp.Partner
  END AS Partner,
  SUM(f.Expenditure) AS Total_Expenditure,
  ROUND(SUM(f.Expenditure) / SUM(SUM(f.Expenditure)) OVER () * 100, 2) AS Percentage_of_Total
FROM mvp.gold.fact_expenditure_partner_year f
JOIN mvp.gold.dim_contributing_partners cp
  ON f.sk_partner = cp.sk_partner
WHERE cp.Partner NOT IN (
    '#BR Rede Globo',
    '#BR Itaipu Binacional',
    '#BR Serviço Social da Indústria',
    'Fundação #BR Renova',
    '#BR UNHCR PSP Brazil',
    '#BR Gerdau',
    '#BR Petrobrás'
)
GROUP BY 
  CASE 
    WHEN cp.Partner = 'ABC - Brazilian Cooperation Agency' 
      OR cp.Partner LIKE '#BR%' 
      OR cp.Partner LIKE 'Brazilian%' 
      THEN 'Self-funded partners'
    ELSE cp.Partner
  END
ORDER BY Total_Expenditure DESC
LIMIT 5;
```
